## Python Functions

There are two kinds of functions in Python.  

-  Built-in functions that are provided as part of Python - `print()`, `input()`, `type()`, `float()`, `int()` ...
-  Functions that we define ourselves and then use.  

We treat function names as “new” reserved words (i.e., we avoid them as variable names) because, like variables, they are **objects** in Python.  

- In Python a function is some reusable code that takes arguments(s) as input, does some computation, and then returns a result or results  
- We define a function using the `def` reserved word  
- We call/invoke the function by using the function name, parentheses, and arguments in an expression

### Using a `docstring` in a **function definition**  

A **docstring** in a Python function is a special string placed at the beginning of the function body, enclosed in triple quotes (""" """ or ''' '''). It describes the function's purpose, parameters, and return value. Docstrings improve code readability and serve as inline documentation.

The `help()` function uses the **docstring** to display information about the function. For example:

```
def add(a, b):
    """Returns the sum of two numbers."""
    return a + b

help(add)  
```

Running `help(add)` prints the **docstring**, making it easier for users to understand the function without checking the source code.

````python
def add(a, b):
    """Returns the sum of two numbers."""
    return a + b

help(add)
````

````python
def add_two_return_three(a, b):
    added = a + b
    print('Added the following two numbers together: ', a, ' and ', b)
    print('Returned the result: ', added)
    return added, a, b
````

## Functions within Functions


Let's play with Fahrenheit to Celsius

````python
def which2which(source_scale, value):
    """ Convert a value from one scale to another """
    def celsius(t_in_f):
        """ returns the temperature in degrees Celsius """
        return ((t_in_f - 32) * 5 / 9)
    def fahrenheit(t_in_c):
        """ returns the temperature in degrees Fahrenheit """
        return ((9 / 5) * t_in_c) + 32
    print("Peeking at my functions")
    print(type(celsius))
    print(type(fahrenheit))
    if source_scale.upper() == 'F':
        return celsius(value)
    elif source_scale.upper() == 'C':
        return fahrenheit(value)
````

````python
c = which2which('f', 32)
print(c)
````
